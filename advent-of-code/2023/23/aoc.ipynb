```python
import pandas as pd
```

```python
inp = pd.read_csv("input23.txt", header=None).to_numpy().flatten()

# get each computer and find which computers it is connected to.
known = {}
for connection in inp:
    c1, c2 = connection.split("-")
    if c1 in known and c2 in known:
        known[c1].append(c2)
        known[c2].append(c1)
    elif c1 in known and c2 not in known:
        known[c1].append(c2)
        known[c2] = [c1]
    elif not c1 in known and c2 in known:
        known[c1] = [c2]
        known[c2].append(c1)
    else: # neither c1 nor c2 in known
        known[c1] = [c2]
        known[c2] = [c1]
# print(known)

# problem part I
# figure out what groups of 3 connected computers exists
groups = []
for comp in known:
    for subComp in known[comp]:
        for subSubComp in known[subComp]:
            if comp in known[subSubComp]:
                # check if the group already exists in groups
                in_group = False
                for group in groups:
                    if (comp in group) and (subComp in group) and (subSubComp in group):
                        in_group = True
                if not in_group:
                    groups.append([comp, subComp, subSubComp])
print(groups)
print(len(groups))

# count the number of groups which have >= 1 computer with a name that starts with t
# count = 0
has_chief = False
chiefGroups = []
for group in groups:
    for member in group:
        if 't' in member[0]:   
            has_chief = True
    if has_chief:
        # count += 1
        chiefGroups.append(group)
        has_chief = False
print(chiefGroups)
print(len(chiefGroups))
# print(count)

# problem part II
groups = []
```
